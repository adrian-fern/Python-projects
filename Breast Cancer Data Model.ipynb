# Breast Cancer Data Models

The code in this section uses pandas and scikit-learn to create a data model that uses various attributes of breast cancer cases to determine if a case is benign or malignant. Two model types (Decision Tree and Logistic Regression) are created and tested with a sample case.

```python
import pandas as pd
from sklearn import tree
from sklearn import metrics
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```
